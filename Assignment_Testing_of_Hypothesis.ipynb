## T -TEST

Two Sample

```python
from scipy.stats import ttest_ind
```

```python
t_stat, p_value = ttest_ind(df['Sales_before_digital_add(in $)'],df['Sales_After_digital_add(in $)'])
```

```python
p_value
```

```python
if p_value<0.05:
  print('Reject Null Hypothesis')
else:
  print('Accept Null Hypothesis')
```

## Chi Square Test

```python
## dependency between "Region" and "Manager"
```

```python
from scipy.stats import chisquare
```

```python
from scipy.stats import chi2_contingency
```

```python
contingency_table = pd.crosstab(df['Region'],df['Manager'])
```

```python
contingency_table
```

```python
chi_stat,p_value,dof,expected = chi2_contingency(contingency_table)
```

```python
p_value
```
